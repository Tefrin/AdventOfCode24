```python
def pingus(report):
    val = 1
    if(len(report) != 1):
        increase = int(report[0]) < int(report[1])
        for i in range(len(report)-1):
            n1 = int(report[i])
            n2 = int(report[i+1])
            val &= not((n1 < n2) ^ increase)
            val &= abs(n1 - n2) >= 1
            val &= abs(n1 - n2) <= 3
    return val
```
